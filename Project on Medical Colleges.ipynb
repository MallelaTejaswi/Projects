```python
a=soup.find_all("div",class_="panel-body")
n=a[0].text.split(",")[1].replace("\n","")
n
```

```python
colleges=[]
for i in a:
    b=i.text
    colleges.append(b.split(",")[0].replace("\n",""))
colleges
```

```python
city=[]
for i in a:
    c=i.text.split(",")[1]
    city.append(re.findall("\s[A-za-z\sA-za-z]+",c)[0].split("\n")[0].replace(" ",""))
city
```

```python
Establish=[]
for i in a:
    d=i.text
    
    
    f=re.findall("\s(\d\d\d\d);",d)
    Establish.append(f[0])
    
Establish
```

```python
Type=[]
for i in a:
    d=i.text
    Type.append(re.findall("([A-Za-z]+);",d)[0])
Type
```

```python
state=[]
st=soup.find_all("div",class_="additional-info")
for i in st:

    state.append(i.text.split(",")[-1].replace(" ",""))
state

```

```python
seats=[]
s=soup.find_all("li",class_="list-group-item")
for i in s:
    seats.append(re.findall("[0-9]+",i.text)[2])
seats
```

```python
fee=[]
fe=soup.find_all('li',class_='list-group-item')
for i in fe:
    fee.append(fe)
```

```python
colleges=[]
Type=[]
Establish=[]
city=[]
seats=[]
state=[]
fee=[]
for i in range(1,19):
    url=f"https://www.indiaccess.com/colleges/medical/index.asp?currentpage={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    a=soup.find_all("div",class_="panel-body")
    s=soup.find_all("li",class_="list-group-item")
    st=soup.find_all("div",class_="additional-info")
    fe=soup.find_all('li',class_='list-group-item')
    
    #for college name 
    for i in a:
        b=i.text
        colleges.append(b.split(",")[0].replace("\n",""))
        
        
    
    #for Type of college name
    for i in a:
        d=i.text
        Type.append(re.findall("([A-Za-z]+);",d)[0])
        
        
        
    #for Established Year    
    for i in a:
        d=i.text
        f=re.findall("\s(\d\d\d\d);",d)
        Establish.append(f[0])
    
    
    
    #for City of the college
    for i in a:
        c=i.text.split(",")[1]
        city.append(re.findall("\s[A-za-z\sA-za-z]+",c)[0].split("\n")[0].replace(" ",""))

    
    
    #for Seats of the college
    for i in s:
        a=i.text
        b=re.findall("Seats:(\s\d+)",a)
        if len(b)>0:
            seats.append(b[0])
    

    
    #for State of the college
    for i in st:
        state.append(i.text.split(",")[-1].replace(" ",""))
    
    

    
print(len(colleges))
print(len(Type))
print(len(Establish))
print(len(city))
print(len(state))

print(len(seats))

        
```

# Building a Data Frame

```python
File=pd.DataFrame({"College":colleges,"Year of establish":Establish,"Type":Type,"City":city,"State":state,"Seats":seats})
```
